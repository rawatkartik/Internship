```python
# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
# = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
# A) Rank 
# B) Name 
# C) Artist 
# D) Upload date 
# E) Views 
```

```python
driver= webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
name= []
artist= []
upload_date= [] 
views= []
```

```python
name_tag= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name_tag:
    song_name= i.text
    name.append(song_name)
    
artist_tag= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist_tag:
    artist_name= i.text
    artist.append(artist_name)
    
upload_date_tag= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in upload_date_tag:
    date= i.text
    upload_date.append(date)
    
views_tag= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in views_tag:
    no_of_views= i.text
    views.append(no_of_views)
```

```python
print(len(name),len(artist),len(upload_date),len(views))
```

```python
df=pd.DataFrame({'Song':name,'Artist':artist,'Upload Date':upload_date,'Views':views})
df
```

```python
# 3. Scrape the details of State-wise GDP of India from statisticstime.com. 
# Url = http://statisticstimes.com/ 
# You have to find following details: A) Rank 
# B) State 
# C) GSDP(18-19)- at current prices 
# D) GSDP(19-20)- at current prices 
# E) Share(18-19) 
# F) GDP($ billion) 
# Note: - From statisticstimes home page you have to reach to economy page through code.
```

```python
driver.get('http://statisticstimes.com/')
```

```python
designation= driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[2]/div[2]/button')
actions = ActionChains(driver)
actions.move_to_element(designation).perform()
```

```python
designation2= driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]')
designation2.click()
```

```python
designation3= driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
designation3.click()
```

```python
ranking= []
states= []
gsdp1= [] 
gsdp2= [] 
share= [] 
gdp= []

rank_tag= driver.find_elements(By.XPATH,'//td[@class="data1"]')
for i in rank_tag:
    ranks= i.text
    ranking.append(ranks)
    
state_tag= driver.find_elements(By.XPATH,'//table[@class="display dataTable"]/tbody/tr/td[2]')
for i in state_tag:
    state= i.text
    states.append(state)

gs_tag= driver.find_elements(By.XPATH,'//table[@class="display dataTable"]/tbody/tr/td[3]')
for i in gs_tag:
    gss= i.text
    gsdp1.append(gss)
    
dp_tag= driver.find_elements(By.XPATH,'//table[@class="display dataTable"]/tbody/tr/td[4]')
for i in dp_tag:
    dps= i.text
    gsdp2.append(dps)
    
share_tag= driver.find_elements(By.XPATH,'//table[@class="display dataTable"]/tbody/tr/td[5]')
for i in share_tag:
    sharetext= i.text
    share.append(sharetext)
    
gdp_tag= driver.find_elements(By.XPATH,"//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp_tag:
    gdps= i.text
    gdp.append(gdps)
    
ranking33= ranking[:33]
states33= states[:33]
gsdp133= gsdp1[:33] 
gsdp233= gsdp2[:33] 
share33= share[:33] 
gdp33= gdp[:33]

print(len(ranking33),len(states33), len(gsdp133), len(gsdp233), len(share33), len(gdp33))
```

```python
df=pd.DataFrame({'Ranking':ranking33,'State Name':states33, 'GSDP(21-22)':gsdp233, 'GSDP(22-23)':gsdp133, 'Share(21-22)':share33, 'GDP':gdp33})
df
```

```python
designation1= driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]')
designation1.click()
```

```python
repositorytitle= []
repositorydescription= []
contributorscount= []
languageused= []
```

```python
title= driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
for i in title:
    if i is None:
        repositorytitle.append('--')
    else:
        repositorytitle.append(i.text)
print(len(repositorytitle))
```

```python
description= driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in range(25):
    if i < len(description):
        repositorydescription.append(description[i].text)
    else:
        repositorydescription.append('--')

count= driver.find_elements(By.XPATH,'//span[@class="d-inline-block float-sm-right"]')
for i in range(25):
    if i < len(count):
        contributorscount.append(count[i].text)
    else:
        contributorscount.append('--')

language_elements = driver.find_elements(By.XPATH, '//span[@itemprop="programmingLanguage"]')
for i in range(25):
    if i < len(language_elements):
        languageused.append(language_elements[i].text)
    else:
        languageused.append('--')
```

```python
print(len(repositorytitle),len(repositorydescription),len(contributorscount),len(languageused))
```

```python
df=pd.DataFrame({'Repository Title':repositorytitle,' Description':repositorydescription,'Contributors':contributorscount,'Language':languageused})
df
```

```python
driver= webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
alldatasets= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
alldatasets.click()
```

```python
expand= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
expand.click()
```

```python
name= []
datatype= [] 
task= []
attribute= [] 
instances= []
no_of_attribute= [] 
year= []
```

```python
name_tag= driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in name_tag:
    names= i.text
    name.append(names)
    
datatype_tag= driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][2]/span[1]')
for i in datatype_tag:
    data= i.text
    datatype.append(data)
    
task_tag= driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][1]/span[1]')
for i in task_tag:
    tasks= i.text
    task.append(tasks)
    
attribute_tag= driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[2]')
for i in attribute_tag:
    attributes= i.text
    attribute.append(attributes)
    
instances_tag= driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][3]/span[1]')
for i in instances_tag:
    instance= i.text
    instances.append(instance)
    
no_of_attribute_tag= driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][4]/span[1]')
for i in no_of_attribute_tag:
    no_of_attributes= i.text
    no_of_attribute.append(no_of_attributes)
    
year_tag= driver.find_elements(By.XPATH,'//tbody[@class="border"]/tr/td[3]')
for i in year_tag:
    years= i.text
    year.append(years)
```

```python
df=pd.DataFrame({'Name':name,'Data type':datatype,'Task':task,'Attribute':attribute,'Instances':instances,'No of Attributes':no_of_attributes,'Year':year})
df
```
