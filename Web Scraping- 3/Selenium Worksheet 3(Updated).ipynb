```python
# 8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
# from any YouTube Video. 
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.youtube.com/watch?v=kJQP7kiw5Fk')
```

```python
comments= []
comment_upvotee= []
timeee= []
```

```python
while len(comments) < 500:
    driver.execute_script("window.scrollTo(0, 100);")
    time.sleep(4)
    comment_tag= driver.find_elements(By.ID,'content-text')
    for i in comment_tag:
        comment= i.text
        comments.append(comment)
        
while len(comment_upvotee) < 500:
    driver.execute_script("window.scrollTo(0, 100);")
    time.sleep(4)
    comment_upvote_tag= driver.find_elements(By.ID,'vote-count-middle')
    for i in comment_upvote_tag:
        comment_upvotes= i.text
        comment_upvotee.append(comment_upvotes)
        
while len(timeee) < 500:
    driver.execute_script("window.scrollTo(0, 100);")
    time.sleep(4)
    timee_tag= driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in timee_tag:
        timeeee= i.text
        timeee.append(timeeee)
```

```python
print(len(comments),len(comment_upvotee),len(timeee))
```

```python
df=pd.DataFrame({'Comments':comments,'Comments Upvote':comment_upvotee,'Posted Time':timeee})
df
```

```python
# 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
# “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
# reviews, privates from price, dorms from price, facilities and property description.
```

```python
driver= webdriver.Chrome()
driver.get("http://www.hostelworld.com/")
```

```python
loc= driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
loc.send_keys('London')
time.sleep(4)
loc1= driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button')
loc1.click()
time.sleep(2)
loc2= driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]/div')
loc2.click()
```

```python
producturls = []

urls = driver.find_elements(By.XPATH, '//a[@class="property-card-container horizontal"]')
for i in urls:
    producturls.append(i.get_attribute("href"))
```

```python
namee= []
distancee= []
ratinge= []
reviewe= []
privatee= []
dorme= []
dese= []
face= []
linke= []
```

```python
for url in producturls:
        driver.get(url)
        time.sleep(1)
        

        try:
            name_tags = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[4]/div/div[1]/h1/div')
            for i in name_tags:
                namee.append(i.text)
        except NoSuchElementException:
            namee.append('-')

        try:
            distance_tags = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div/div[4]/div[1]/a/a/div[2]/div[2]/span[2]')
            for i in distance_tags:
                distancee.append(i.text)
        except NoSuchElementException:
            distancee.append('-')

        try:
            rating_tags = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[1]')
            for i in rating_tags:
                ratinge.append(i.text)
        except NoSuchElementException:
            ratinge.append('-')

        try:
            review_tags = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[2]/div[1]')
            for i in review_tags:
                reviewe.append(i.text)
        except NoSuchElementException:
            reviewe.append('-')
            
        try:
            des_tags = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[2]/div[3]/div/div[2]')
            for i in des_tags:
                dese.append(i.text)
        except NoSuchElementException:
            dese.append('-')
```

```python
print(len(namee),len(distancee),len(ratinge),len(reviewe),len(producturls))
```

```python
df=pd.DataFrame({'Name':namee,'Distance from Centre':distancee,'Ratings':ratinge,'Reviews':reviewe,'URL':producturls})
df
```
